```python
def process_dataframe_add_search_column(df):
    """
    Process a dataframe by adding a 'search_column' that contains all unique words
    from 'book name', 'title', and 'publisher' columns.

    Parameters:
    -----------
    df : pandas DataFrame
        The input DataFrame with at least some of the columns: 'book name', 'title', 'publisher'

    Returns:
    --------
    pandas.DataFrame
        DataFrame with the new 'search_column' added
    """
    # Create a copy of the dataframe to avoid modifying the original
    result_df = df.copy()

    # Define columns to process
    columns_to_combine = ['book name', 'title', 'publisher']

    # Filter to only include columns that exist in the dataframe
    existing_columns = [col for col in columns_to_combine if col in df.columns]

    if not existing_columns:
        raise ValueError("None of the required columns found in the dataframe")

    # Function to process each row
    def combine_unique_words(row):
        # Combine text from all specified columns, handling potential NaN values
        combined_text = ' '.join([str(row[col]) for col in existing_columns if pd.notna(row[col])])

        # Convert to lowercase and split into words
        words = combined_text.lower().split()

        # Remove punctuation from words
        words = [re.sub(r'[^\w\s]', '', word) for word in words]

        # Get unique words only (preserving order of first appearance)
        unique_words = []
        for word in words:
            if word and word not in unique_words:  # Skip empty strings
                unique_words.append(word)

        # Join the unique words back into a string
        return ' '.join(unique_words)

    # Apply the function to create the new column
    result_df['search_column'] = result_df.apply(combine_unique_words, axis=1)

    return result_df



processed_df = process_dataframe_add_search_column(df)
processed_df.head()
```

```python
import pandas as pd
import numpy as np
from nltk.metrics.distance import edit_distance

def find_matches_between_dataframes(yivo_books_df, processed_df, similarity_threshold=0.8):
    """
    Find matches between strings in yivo_books_df['vertitle'] and processed_df['search_column']
    based on word similarity (80% of the letters are the same).

    Args:
        yivo_books_df: DataFrame containing 'vertitle' column
        processed_df: DataFrame containing 'search_column' column
        similarity_threshold: Minimum similarity threshold for words to be considered similar (default: 0.8)

    Returns:
        DataFrame with matching results
    """
    # Define stop words for English and Hebrew
    english_stop_words = {
        'a', 'an', 'the', 'and', 'or', 'but', 'in', 'on', 'at', 'to',
        'of', 'for', 'with', 'by', 'about', 'as', 'into', 'like', 'through',
        'after', 'over', 'between', 'out', 'against', 'during', 'without',
        'is', 'are', 'was', 'were', 'be', 'been', 'being',
        'i', 'me', 'my', 'mine', 'myself', 'you', 'your', 'yours', 'yourself',
        'he', 'him', 'his', 'himself', 'she', 'her', 'hers', 'herself',
        'it', 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves'
    }

    hebrew_stop_words = {
        'של', 'את', 'עם', 'על', 'אל', 'מן', 'בין', 'כי', 'אם', 'או',
        'גם', 'רק', 'אך', 'אבל', 'אז', 'כך', 'לכן', 'עוד', 'כבר', 'שם',
        'אני', 'אתה', 'היא', 'הוא', 'אנחנו', 'אתם', 'הן', 'הם'
    }

    stop_words = english_stop_words.union(hebrew_stop_words)

    def calculate_word_similarity(word1, word2):
        """
        Calculate similarity between two words using edit distance.

        Args:
            word1: First word
            word2: Second word

        Returns:
            Similarity score between 0 and 1
        """
        # Convert to lowercase for case-insensitive comparison
        w1 = word1.lower()
        w2 = word2.lower()

        # If words are identical, return 1.0
        if w1 == w2:
            return 1.0

        # Calculate edit distance
        distance = edit_distance(w1, w2)
        max_length = max(len(w1), len(w2))

        # Calculate similarity as 1 - (distance / max_length)
        return 1 - (distance / max_length)

    def remove_stop_words(text):
        """
        Remove stop words from a text string.

        Args:
            text: Input text string

        Returns:
            List of words without stop words
        """
        words = text.lower().split()
        return [word for word in words if word not in stop_words]

    def find_best_match(search_string, vertitle_series):
        """
        Find the best match for a search string from a series of vertitles.

        Args:
            search_string: String to search for
            vertitle_series: Series of vertitle strings

        Returns:
            Tuple of (best_match, score)
        """
        search_words = remove_stop_words(search_string)

        if not search_words:
            return None, 0.0

        best_match = None
        best_score = 0.0

        for vertitle in vertitle_series:
            if pd.isna(vertitle):
                continue

            vertitle_words = remove_stop_words(vertitle)
            matching_word_count = 0

            # For each word in search string, find best match in vertitle
            for search_word in search_words:
                best_word_similarity = 0.0

                for vertitle_word in vertitle_words:
                    similarity = calculate_word_similarity(search_word, vertitle_word)
                    best_word_similarity = max(best_word_similarity, similarity)

                # If similarity is above threshold, count as a match
                if best_word_similarity >= similarity_threshold:
                    matching_word_count += 1

            # Calculate match score as proportion of matching words
            match_score = matching_word_count / len(search_words) if search_words else 0.0

            # Update best match if score is higher
            if match_score > best_score:
                best_score = match_score
                best_match = vertitle

        return best_match, best_score

    # Create results DataFrame
    results = []

    # Convert vertitle column to a list for better performance
    vertitle_list = yivo_books_df['vertitle'].tolist()

    # Process each search string
    for _, row in processed_df.iterrows():
        search_string = row['book name']

        if pd.isna(search_string):
            results.append({
                'search_string': search_string,
                'best_match': None,
                'similarity_score': 0.0
            })
            continue

        best_match, score = find_best_match(search_string, vertitle_list)

        results.append({
            'search_string': search_string,
            'best_match': best_match,
            'similarity_score': score
        })

    return pd.DataFrame(results)

matches_df = find_matches_between_dataframes(yivo_books, df)
# # Save results to CSV
matches_df.to_csv('/Users/noamkashi/Documents/Strashum Library/yivo_for_the_last_time/second_test.csv', index=False)
```
